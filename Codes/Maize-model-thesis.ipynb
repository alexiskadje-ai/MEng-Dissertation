```python
def load_and_preprocess_data():
    import pandas as pd
    import os
    from chrono import Date
    # Loading data
    df = pd.read_csv(r'C:\Users\Alex KADJE\Desktop\MEng Thesis\Thesis_data.csv')
    
    # Cleaning headers and values
    df.columns = df.columns.str.strip().str.replace('"', '')
    df = df.apply(lambda x: x.str.strip().str.replace('"', '') if x.dtype == "object" else x)
    

    
    # Converting dates to datetime
    df['Planting_Date'] = pd.to_datetime(df['Planting_Date'], errors='coerce')
    df['Harvest_Date'] = pd.to_datetime(df['Harvest_Date'], errors='coerce')

    
    # Calculating growing period (in days)
    df['Growing_Period_Days'] = (df['Harvest_Date'] - df['Planting_Date']).dt.days
    
    # Missing value imputation
    numeric_cols = ['Rainfall_mm', 'Avg_Temp_C', 'Soil_pH', 'Organic_Matter_pct', 
                    'Area_Harvested_ha', 'Yield_tons_ha', 'Nitrogen_mg_kg', 
                    'Phosphorus_mg_kg', 'Potassium_mg_kg', 'Humidity_pct', 
                    'Solar_Radiation_W_m2', 'Wind_Speed_m_s', 'Soil_Moisture_pct', 
                    'Fertilizer_kg_ha', 'Pest_Incidents', 'Market_Price_USD_ton', 
                    'Farmer_Input_Cost_USD']
    categorical_cols = ['Irrigation_Method', 'Soil_Type', 'Fertilizer_Use']
    
    for col in numeric_cols:
     if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')  # convert entire column to numeric first
        df[col] = df.groupby('Region')[col].transform(
            lambda x: x.fillna(x.median())
        )

    
    for col in categorical_cols:
        if col in df.columns:
            df[col] = df.groupby('Region')[col].transform(lambda x: x.fillna(x.mode().iloc[0] if not x.mode().empty else 'Unknown'))
    
    # Dropping rows with missing key predictors
    key_predictors = ['Yield_tons_ha', 'Rainfall_mm']
    df = df.dropna(subset=key_predictors)
    
    # Removing duplicates
    df = df.drop_duplicates()
    
    return df
```

```python
# Feature engineering
def feature_engineering(df):
    # Calculating Growing Degree Days (GDD)
    base_temp = 10  # Base temperature for maize
    df['GDD'] = df['Avg_Temp_C'].apply(lambda x: max(0, x - base_temp) if pd.notnull(x) else 0) * df['Growing_Period_Days']
    
    # Nitrogen Mineralization Potential
    df['N_min'] = 0.8 * df['Organic_Matter_pct'] * np.exp(0.07 * df['Avg_Temp_C'])
    
    # Simulating bi-weekly NDVI and Rainfall sequences
    n_snapshots = 12  # Bi-weekly over ~133 days
    df['NDVI_seq'] = df['NDVI'].apply(lambda x: np.linspace(x if pd.notnull(x) else 0.7, x if pd.notnull(x) else 0.7, n_snapshots))
    df['Rainfall_seq'] = df['Rainfall_mm'].apply(lambda x: np.linspace(x/n_snapshots, x/n_snapshots, n_snapshots))
    
    return df
```

```python
# Preparing data for ConvLSTM and XGBoost
def prepare_model_data(df):
    # Features for ConvLSTM (spatiotemporal)
    spatio_temporal_features = ['NDVI_seq', 'Rainfall_seq']
    static_features = ['Soil_pH', 'Organic_Matter_pct', 'Nitrogen_mg_kg', 
                      'Phosphorus_mg_kg', 'Potassium_mg_kg', 'Fertilizer_kg_ha', 
                      'Pest_Incidents', 'Avg_Temp_C', 'GDD', 'N_min']
    
    # Preparing spatiotemporal data (12x1x1x2 tensor for each sample)
    X_spatio = np.array([np.stack([row['NDVI_seq'], row['Rainfall_seq']], axis=-1) for _, row in df.iterrows()])
    X_spatio = X_spatio.reshape(X_spatio.shape[0], 12, 1, 1, 2)
    
    # Preparing static data
    X_static = df[static_features].fillna(0).values
    
    # Target variable
    y = df['Yield_tons_ha'].values
    
    return X_spatio, X_static, y, static_features
```

```python
# Attention mechanism layer
def attention_mechanism(inputs):
    convlstm_out, static_out = inputs
    # Computing attention weights
    concat = Concatenate()([convlstm_out, static_out])
    attention = Dense(1, activation='tanh')(concat)
    attention = Dense(1, activation='softmax')(attention)
    # Applying attention weights
    weighted_convlstm = Lambda(lambda x: x[0] * x[1])([convlstm_out, attention])
    weighted_static = Lambda(lambda x: x[0] * x[1])([static_out, 1 - attention])
    return Concatenate()([weighted_convlstm, weighted_static])
```

```python
# Building the ConvLSTM-XGBoost hybrid model with attention
def build_hybrid_model():
    # ConvLSTM branch
    spatio_input = Input(shape=(12, 1, 1, 2))
    x = ConvLSTM2D(filters=32, kernel_size=(1, 1), padding='same', return_sequences=True)(spatio_input)
    x = ConvLSTM2D(filters=64, kernel_size=(1, 1), padding='same')(x)
    convlstm_out = Flatten()(x)
    
    # Static input for XGBoost features
    static_input = Input(shape=(10,))
    static_out = Dense(64, activation='relu')(static_input)
    
    # Attention mechanism
    combined = attention_mechanism([convlstm_out, static_out])
    
    # Fusion layer
    x = Dense(64, activation='relu')(combined)
    output = Dense(1)(x)
    
    # Keras model for ConvLSTM with attention
    model = Model(inputs=[spatio_input, static_input], outputs=output)
    model.compile(optimizer='adam', loss='mse')
    
    # XGBoost model
    xgb_model = xgb.XGBRegressor(n_estimators=100, max_depth=6, learning_rate=0.1)
    
    return model, xgb_model
```

```python
# SHAP explanation
def explain_model(xgb_model, X_static, static_features):
    explainer = shap.TreeExplainer(xgb_model)
    shap_values = explainer.shap_values(X_static)
    
    # Plotting SHAP summary
    shap.summary_plot(shap_values, X_static, feature_names=static_features, show=False)
    plt.savefig('shap_summary_plot.png')
    plt.close()
```

```python
# Training and evaluating the model
def train_and_evaluate():
    df = load_and_preprocess_data()
    df = feature_engineering(df)
    X_spatio, X_static, y, static_features = prepare_model_data(df)
    
    model, xgb_model = build_hybrid_model()
    
    # K-Fold Cross-Validation
    kf = KFold(n_splits=5, shuffle=True, random_state=42)
    mae_scores = []
    r2_scores = []
    
    for train_index, test_index in kf.split(X_spatio):
        X_spatio_train, X_spatio_test = X_spatio[train_index], X_spatio[test_index]
        X_static_train, X_static_test = X_static[train_index], X_static[test_index]
        y_train, y_test = y[train_index], y[test_index]
        
        # Training ConvLSTM with attention
        model.fit([X_spatio_train, X_static_train], y_train, epochs=100, batch_size=16, verbose=0)
        
        # Training XGBoost
        xgb_model.fit(X_static_train, y_train)
        
        # Predicting with ConvLSTM
        y_pred_conv = model.predict([X_spatio_test, X_static_test], verbose=0)
        
        # Combining predictions (simple averaging for fusion)
        y_pred_xgb = xgb_model.predict(X_static_test)
        y_pred = 0.5 * y_pred_conv.flatten() + 0.5 * y_pred_xgb
        
        # Evaluating
        mae = mean_absolute_error(y_test, y_pred)
        r2 = r2_score(y_test, y_pred)
        mae_scores.append(mae)
        r2_scores.append(r2)
    
    print(f"Average MAE: {np.mean(mae_scores):.3f} tons/ha")
    print(f"Average R: {np.mean(r2_scores):.3f}")
    
    # SHAP explanation for XGBoost
    explain_model(xgb_model, X_static, static_features)

if __name__ == "__main__":
    train_and_evaluate()
```
